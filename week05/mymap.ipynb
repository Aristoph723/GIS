```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
reitaku = [35.8330, 139.9550]
home = [22.31714268320589, 114.17460248908395]
vic = [22.3045819135081, 114.16633571420509]
mrg = [22.316074325400447, 114.17270248928038]
npl = [22.312985909216867, 114.19014453969243]

# make the map
m = folium.Map(location=home, zoom_start=15)
r = folium.Map(location=reitaku, zoom_start=15)

# add a marker
folium.Marker(home, popup="My home, there's a supermarket right in front of our building!", tooltip="My Home").add_to(m)
folium.Marker(vic, popup="Hong Kong's ever-popular night time sight-seeing spot!", tooltip="Victoria Harbour").add_to(m)
folium.Marker(mrg, popup="My go-to Gunpla store. Over a decade it has had 2 different owners",tooltip="Mr.Gundam").add_to(m)
folium.Marker(npl,popup="My Secondary School! I played Volleyball for 6 years here",tooltip="Po Leung Kuk Ngan Po Ling College").add_to(m)
folium.Marker(reitaku, popup="This is Reitaku University, I study here!", tooltip="Reitaku University").add_to(r)

# show the map
m
```

```python
m.save("ymt_map.html")
```
